```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--no-sandbox")
# options.add_argument("--headless")
# options.add_argument("--disable-gpu")
options.add_argument("--disable-dve-shm-usage")
```

```python
driver = webdriver.Chrome(options=options)
driver.maximize_window()
```

```python
driver.get("https://www.google.com")
```

```python
temp=[]
```

```python
def Google_Searching(key): 
        search = WebDriverWait(driver, 50).until(
                EC.visibility_of_element_located((By.CSS_SELECTOR,"#APjFqb")))
        search.send_keys(key)
        search.send_keys(Keys.RETURN)
        button = WebDriverWait(driver, 50).until(
                EC.visibility_of_element_located((By.CSS_SELECTOR,"#tsf > div:nth-child(1) > div.A8SBwf > div.RNNXgb > button > div > span > svg")))
        button.click()
#         button = driver.find_element(By.CSS_SELECTOR,"#tsf > div:nth-child(1) > div.A8SBwf > div.RNNXgb > button > div > span > svg").click()
# #     Bangla_to_English = driver.find_element(By.CSS_SELECTOR,"#Rzn5id > div > a:nth-child(3)")
# #     Bangla_to_English.click()
# #     time.sleep(1)
    
        urls = driver.find_elements(By.CSS_SELECTOR," a[jsname='UWckNb']")
        for i in urls:
            url = i.get_attribute("href")
            if ".gov.uk" in url:
                continue
#             print(url)
            else:
                url = url
                dic = {
                    "Title":key,
                    "website":url
                }
                temp.append(dic)
                break
#         close = driver.find_element(By.CSS_SELECTOR,"#tsf > div:nth-child(1) > div.A8SBwf > div.RNNXgb > div > div.dRYYxd > div.BKRPef > div > span > svg").click()
        close = WebDriverWait(driver, 50).until(
                EC.visibility_of_element_located((By.CSS_SELECTOR,"#tsf > div:nth-child(1) > div.A8SBwf > div.RNNXgb > div > div.dRYYxd > div.BKRPef > div > span > svg"))).click()    
        time.sleep(2)
        
```
