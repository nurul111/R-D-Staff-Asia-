```python
ttt = 0

for page in range(1, 101): 
    print(f"Running page: {page}")
    pagination_string = f"&page={page}"
    
    new_url = url + pagination_string

    driver.get(new_url)
    
    time.sleep(8)
    
    
    all_results = driver.find_elements(By.CSS_SELECTOR, 'tbody.zp_RFed0')
    print(f"in this page: {len(all_results)}")
    print("-------------------------------")
    for company in all_results[0:]:
        m = {}

        try: 
            person_name = company.find_element(By.CSS_SELECTOR, "div.zp_BC5Bd").text
        except:
            person_name = "N/A"
        try: 
            job_title = company.find_element(By.CSS_SELECTOR, "tr > td:nth-child(2) > span").text
        except:
            job_title = "N/A"
        try: 
            main_company = company.find_element(By.CSS_SELECTOR, "div.zp_J1j17 > a").text
        except:
            main_company = "N/A"
        try:
            loc = company.find_element(By.CSS_SELECTOR, " tr > td:nth-child(6) > div:nth-child(1) > div > div > div.zp_DBwlj > span").text
        except:
            try:
                loc = company.find_element(By.CSS_SELECTOR, "tr > td:nth-child(6) > span").text
            except:
                loc="N/A"
        try:
            ind = company.find_element(By.CSS_SELECTOR,"tr > td:nth-child(8) > span > div").text
        except:
            ind = "N/A"
       
        try:
            social_links = company.find_elements(By.CSS_SELECTOR, 'tr > td:nth-child(3) > div:nth-child(1) > div > div > div.zp_DBwlj > span > div > div.zp_I1ps2 > span > a')
        except:
            try:
                social_links = company.find_elements(By.CSS_SELECTOR, "tr > td:nth-child(3) > span > div > div.zp_I1ps2 > span > a")
            except:
                social_links  = []

        company_linkedin_link = "N/A"
        company_fb_link = "N/A"
        company_twtr_link = "N/A"
        company_website_link = "N/A"

        for link in social_links:

            tmp_link = link.get_attribute('href')

            if 'linkedin' in tmp_link:
                company_linkedin_link = tmp_link
            elif 'facebook' in tmp_link:
                company_fb_link = tmp_link
            elif 'twitter' in tmp_link:
                company_twtr_link = tmp_link
            else:
                company_website_link = tmp_link
        print(company_website_link)
        time.sleep(1)
        email_butt = company.find_element(By.CSS_SELECTOR, "tr >  td:nth-child(4) i.zp-icon.apollo-icon.apollo-colored-icon.zp_dZ0gM.zp_j49HX.zp_uAV5p").click()
        time.sleep(10)
        try:
            person_email = driver.find_element(By.CSS_SELECTOR," div.zp_JywRU > span").text
            email_type = driver.find_element(By.CSS_SELECTOR," div.zp_DMnwE").text
            email_status = driver.find_element(By.CSS_SELECTOR," div.zp_Q3JNy.zp_On5wr").text
        except:
            person_email = "N/A"
            email_type = "N/A"
            email_status = "N/A"
        time.sleep(0.5)
        try:
            email_butt = company.find_element(By.CSS_SELECTOR, "tr >  td:nth-child(4) i.zp-icon.apollo-icon.apollo-colored-icon.zp_dZ0gM.zp_j49HX.zp_uAV5p").click()
        except:
            email_butt = company.find_element(By.CSS_SELECTOR, "#contactFinderActions-640eff6388795e0001edd944 > div > div > button.zp-button.zp_zUY3r.zp_hLUWg.zp_n9QPr.zp_B5hnZ.zp_MCSwB.zp_IYteB").click()
        time.sleep(2)
        
        #####
        try: 
            p = company.find_element(By.CSS_SELECTOR, "tr > td:nth-child(7) > span")
            p.click()

            try:
                key = WebDriverWait(driver, 5).until(
                    EC.visibility_of_element_located((By.CSS_SELECTOR, "div.zp_ZMABY > div.zp_krQXB")) 
                )
                keys = key.text
            except:
                keys = "N/A"
          
        except:
            pass
        #             print(person_email)
        try:
            time.sleep(2)
            back_button = driver.find_element(By.CSS_SELECTOR, 'div.zp_IrSry.zp_v_0R3')
            back_button.click()
            time.sleep(2)
        except:
            pass
            
        
        m = {
            'person_name': person_name,
            'job_title': job_title,
            'location': loc,
            'Industry':ind,
            'Keyword':keys,
            'main_company' : main_company,
            'person_email': person_email,
            'email_type': email_type,
            'email_status':email_status,
            'company_website' : company_website_link, 
            'company_linkedin_link' : company_linkedin_link,
            'company_fb_link': company_fb_link,
            'company_twtr_link': company_twtr_link,
            }

        data.append(m)
```
