```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--no-sandbox")
# options.add_argument("--headless")
# options.add_argument("--disable-gpu")
options.add_argument("--disable-dve-shm-usage")
```

```python
driver = webdriver.Chrome(options=options)
driver.maximize_window()
```

```python
driver.get("https://www.managers.org.uk/education-and-learning/find-a-centre/")
time.sleep(5)
```

```python
iframe = driver.find_element(By.CSS_SELECTOR, '#main > div > div.fl-row.fl-row-full-width.fl-row-bg-none.fl-node-5ea0196aec8ef.fl-row-default-height.fl-row-align-center > div > div > div > div > div > div > div > div > iframe')  # Replace with the actual iframe id or XPath
driver.switch_to.frame(iframe)
```

```python
centers = driver.find_element(By.CSS_SELECTOR,"#pnlMapCentres > div > div > div:nth-child(3) > div:nth-child(3) > a")
print(centers.text)
```

```python
centers = driver.find_elements(By.CSS_SELECTOR,"#pnlMapCentres > div > div > div")
```

```python
print(len(centers))
```

```python
data = []
```

```python
count = 361
for center in centers[361:]:
    print(count)
#     print(center.text)
    button = center.find_element(By.CSS_SELECTOR,"div > a")
#     print(button.text)
    time.sleep(2)
    button.click()
    time.sleep(5)
    try:
        head = driver.find_element(By.CSS_SELECTOR,"#myModalLabel").text
        if head == " " or head=="":
            head=driver.find_element(By.CSS_SELECTOR,"#myModalLabel").text
    except:
        head = "N/A"
    try:
        adrs = driver.find_element(By.CSS_SELECTOR,"#pAddress").text
        if adrs == " " or adrs=="":
            adrs="N/A"
    except:
        adrs = "N/A"
    try:
        town = driver.find_element(By.CSS_SELECTOR,"#pTown").text
        if town == " " or town=="":
            town="N/A"
    except:
        town = "N/A"
    try:
        cou = driver.find_element(By.CSS_SELECTOR,"#pCounty").text
        if cou == " " or cou=="":
            cou="N/A"
    except:
        cou = "N/A"
    try:
        pc = driver.find_element(By.CSS_SELECTOR,"#pPostCode").text
        if pc == " " or pc=="":
            pc="N/A"
    except:
        pc = "N/A"
    try:
        phone = driver.find_element(By.CSS_SELECTOR,"#pPhone").text
        if phone == " " or phone=="":
            phone="N/A"
    except:
        phone = "N/A"
    try:
        email = driver.find_element(By.CSS_SELECTOR,"#pEmail").text
        if email == " " or email=="":
            email="N/A"
    except:
        email = "N/A"
    try: 
        website = driver.find_element(By.CSS_SELECTOR,"#pWebsite").text
        if website == " " or website=="":
            website="N/A"
    except:
        website = "N/A"
    try:
        Bio = driver.find_element(By.CSS_SELECTOR,"#pModalBio").text
        if Bio == " " or Bio=="":
            Bio="N/A"
    except:
        Bio = "N/A"
    
    
    qualf = driver.find_elements(By.CSS_SELECTOR,"#divQuals p")
        
    qualification = []
    for i in qualf:
        qualification.append(i.text)
        
        
    dic = {
        "Title":head,
        "street_Address":adrs,
        "Town":town,
        "County":cou,
        "PostCode":pc,
        "Phone":phone,
        "Email":email,
        "website":website,
        "Qualification":qualification,
        "BIO":Bio
    }
    close = driver.find_element(By.CSS_SELECTOR,"#centreModal > div > div > div.modal-header > button")
    time.sleep(1)
    close.click()
    data.append(dic)
    count+=1
```

```python
df = pd.DataFrame(data)

df['Phone'] = '`' + df['Phone']
df
```

```python
df.to_csv("../managers.org.uk/Education and Learning(center) new.csv")
```
